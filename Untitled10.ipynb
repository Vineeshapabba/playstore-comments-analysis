```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf=Pipeline([
    ('countVec',CountVectorizer()),
    ('model',MultinomialNB())
])
```

```python
clf.fit(df['content'],df['score'])
```

```python
clf.predict(["I've been playing this game for over a year and I've had a great experience, graphics are very good, it has many tracks and cars to play with, it's absolutely best android racing game but i'm rating it 4 stars because this game need a mode that i like very much, in which we can play in any map with any car solo to try out new cars, set track records, beat our best time etc. If the mode gets added to the game this will absolutely be a 10/10."])
```
